```python
#C:\Users\abdel\Desktop\Year 4 Indiv. project\dataset_of_sounds\PlantSounds\Tomato Dry

import os
import scipy.io.wavfile as wavfile
import numpy as np
import pandas as pd
from scipy.signal import find_peaks
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns   
```

```python


def perform_fft(file_path):
    # Read the wav file
    sample_rate, data = wavfile.read(file_path)

    # Perform FFT
    fft_result = np.fft.rfft(data)
    fft_freq = np.fft.rfftfreq(len(data), 1/sample_rate)

    # Find the peak frequency
    peak, _ = find_peaks(np.abs(fft_result), distance=sample_rate/2)
    max_freq = fft_freq[peak]

    # Take the maximum frequency if there are multiple peaks
    if max_freq.size > 0:
        max_freq = np.max(max_freq)
    else:
        max_freq = np.nan

    # Calculate the half power
    half_power = np.max(np.abs(fft_result)) / np.sqrt(2)

    # Find the first frequency above max_freq that reaches half_power
    indices_above = np.where((fft_freq > max_freq) & (np.abs(fft_result) >= half_power))
    freq_above = np.min(fft_freq[indices_above]) if indices_above[0].size > 0 else np.nan

    # Find the first frequency below max_freq that reaches half_power
    indices_below = np.where((fft_freq < max_freq) & (np.abs(fft_result) >= half_power))
    freq_below = np.max(fft_freq[indices_below]) if indices_below[0].size > 0 else np.nan

    # Calculate the bandwidth
    bandwidth = freq_above - freq_below if freq_above is not np.nan and freq_below is not np.nan else np.nan

    # Calculate the spectral centroid
    spectral_centroid = np.sum(fft_freq * np.abs(fft_result)) / np.sum(np.abs(fft_result))

    # # Calculate the energy entropy
    # frame_size = 1  # Each sample is treated as a frame
    # num_frames = len(data)  # Total number of frames is equal to total number of samples
    # energies = np.array([np.square(data[i]) for i in range(num_frames)])  # Energy of each frame
    # energies_normalized = energies / np.sum(energies)
    # epsilon = 1e-10  # Small constant to avoid taking log2 of 0
    # lg = np.log2(energies_normalized + epsilon)
    # energy_entropy = -np.sum(energies_normalized * np.log10(energies_normalized + epsilon))
    # print("energies",energies_normalized)
    energy_entropy = stats.entropy(np.abs(data))

    return max_freq, bandwidth, spectral_centroid, energy_entropy

# Directory containing the .wav files
directory = 'C:/Users/abdel/Desktop/Year 4 Indiv. project/dataset_of_sounds/PlantSounds/Tomato Dry'

# Get a list of all .wav files in the directory
wav_files = [f for f in os.listdir(directory) if f.endswith('.wav')]

# Initialize an empty DataFrame to store the results
df = pd.DataFrame(columns=['File', 'Max Frequency', 'Bandwidth', 'Spectral Centroid', 'Energy Entropy'])

# Loop over the .wav files
for wav_file in wav_files:
    # Perform FFT on the wav file
    max_freq, bandwidth, spectral_centroid, energy_entropy = perform_fft(os.path.join(directory, wav_file))
    #print(type(max_freq))
    # Append the results to the DataFrame
    df = df.append({'File': wav_file,
                    'Max Frequency': max_freq,
                    'Bandwidth': bandwidth,
                    'Spectral Centroid': spectral_centroid, 
                    'Energy Entropy': energy_entropy}, ignore_index=True)

```

```python
## intesity of sounds
def intensity_analysis(file_path):
    # Read the wav file
    sample_rate, data = wavfile.read(file_path)

    # Perform FFT
    fft_result = np.fft.rfft(data)
    fft_freq = np.fft.rfftfreq(len(data), 1/sample_rate)

    # Find the peak frequency
    peak, _ = find_peaks(np.abs(fft_result), distance=sample_rate/2)
    max_freq = fft_freq[peak]

    # Take the maximum frequency if there are multiple peaks
    if max_freq.size > 0:
        max_freq = np.max(max_freq)
    else:
        max_freq = np.nan

    #max intensity calculation 
    # Convert all values in data to absolute values
    data_abs = np.abs(data)

    # Find the maximum value (intensity)
    intensitybit = np.max(data_abs)

    return max_freq, intensitybit

# Directory containing the .wav files
directory = 'C:/Users/abdel/Desktop/Year 4 Indiv. project/dataset_of_sounds/PlantSounds/Tomato Dry'

# Get a list of all .wav files in the directory
wav_files = [f for f in os.listdir(directory) if f.endswith('.wav')]

# Initialize an empty DataFrame to store the results
df = pd.DataFrame(columns=['File', 'Max Frequency', 'intensity'])

# Loop over the .wav files
for wav_file in wav_files:
    # Perform FFT on the wav file
    #max_freq, bandwidth, spectral_centroid, energy_entropy = perform_fft(os.path.join(directory, wav_file))
    max_freq, intensitybit = intensity_analysis(os.path.join(directory, wav_file))
    #print(type(max_freq))
    # Append the results to the DataFrame
    df = df.append({'File': wav_file,
                    'Max Frequency': max_freq,
                    'intensitybit': intensitybit}, ignore_index=True)
    

```
